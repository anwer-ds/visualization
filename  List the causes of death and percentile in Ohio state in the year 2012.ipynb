#### 1: List the causes of death and percentile in Ohio state in the year 2012

```python
#import required libraries

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# View the plot 
```

```python
#cause of deaths
cause='Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis','Others'

percentile=[62, 5, 4, 2, 1 , 26 ]
#percentile

```

#### 2: Draw a pie chart using the two data points

```python
#set the pie chart plot properties
```

```python
#set figure size
plt.figure(figsize=(5,5))
explode=(0.01,0,0,0,0,0)
plt.pie(percentile,labels=cause,explode=explode,autopct='%1.8f%%',startangle=70)
plt.axis('equal')
plt.title(' List the causes of death and percentile in Ohio state in the year 2012')
plt.show()
```
